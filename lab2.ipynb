Из двух массивов сделать матрицу Коши C (Cij = 1/(xi - yj))

```python
x=np.arange(4)
y=x+6
z=np.ones((4,4))
print(x)
print(y)
for i in range(0,len(x)):
    for j in range(0,len(y)):
        z[i,j]=1/(x[i]-y[j])
print(z)        
```

Определить, есть ли в 2D массиве нулевые столбцы

```python
x=np.array([[1,0,2],[0,0,3],[1,0,9]])
for i in range(0,3):
    n=0
    for j in range(0,3):
        if (x[j,i]==0):
            n=n+1
    if (n==3):
        print("В матрице есть нулевой столбец")
        break
if (n!=3):
    print("В матрице нет нулевых столбцов")
        
        
```

Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве (осторожно с повторами)

```python
x=np.random.randint(0,10,10)
print(x)
y=np.random.randint(0,len(x)-1,20)
print(y)
z=np.unique(y)
print(z)
for i in range(0,len(z)):
    x[z[i]]=x[z[i]]+1
print(x)    
```
