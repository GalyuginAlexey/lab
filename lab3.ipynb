Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

```python
x=np.array([1,2,3,4,5])
y=np.ones(len(x)*4)
for i in range(0,len(x)*4):
    if (i%4==0):
        y[i]=x[i//4]
    else:
        y[i]=0
print(y)        
```

Дан массив 16x16, посчитать сумму по блокам 4x4

```python
global x
x=np.ones((16,16))
def dot(inach,ikont,jnach,jkont):
    count=0
    for i in range(inach,ikont):
        for j in range(jnach,jkont):
            count=count+x[i,j]
    return(count)
for i in range(0,16,4):
    j=i+4
    for k in range(0,16,4):
        m=k+4
        print(dot(i,j,k,m))
```

Построить прямое произведение массивов (все комбинации с каждым элементом)

```python
def dot(arrays):
    arrays=[np.asarray(a) for a in arrays]
    shape=map(len,arrays)
    ix=np.indices(shape,dtype=int)
    ix=ix.reshape(len(arrays),-1).T
    for n, arr in enumerate(arrays):
        ix[:,n]=arrays[n][ix[:,n]]
    return ix
print(dot(([1, 2, 3], [4, 5])))
```
